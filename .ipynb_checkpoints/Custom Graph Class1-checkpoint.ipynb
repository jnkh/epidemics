```julia
using PyPlot
```

```julia
#include("GraphModules.jl")
using SIS
using IM
import Graphs

function run_epidemics(N::Int,k::Int,num_runs::Int,im::InfectionModel)  
    runs = {}
    for i in 1:num_runs
        g = Graphs.erdos_renyi_graph(N,1.0*k/(N-1),is_directed=false)
        t = get_type_graph(g,SUSCEPTIBLE)
        infecteds = {}
        set_type(Graphs.vertices(t)[1],INFECTED)
        frac = get_fraction_of_type(t,INFECTED)
        push!(infecteds,frac)

        while frac > 0 && frac < 1
            update_graph(t,im)
            frac = get_fraction_of_type(t,INFECTED)
            push!(infecteds,N*frac)
            print("\r",N*frac, " ")
            flush(STDOUT)
        end
        push!(runs,infecteds)
    end

    num_fixed = 0
    sizes = zeros(num_runs)
    total_length =0

    for (i,run) in enumerate(runs)
        if run[end] == 1.0
            num_fixed += 1
            sizes[i] = Inf
        end
        sizes[i] = im.dt*sum(run)
        total_length += length(run)
    end
    #get rid of fixed ones
    sizes = sizes[sizes .< Inf]

    return sizes,num_fixed,total_length
end
```

```julia
bins = logspace(log10(minimum(sizes)),log10(maximum(sizes)),40)
PyPlot.plt.hist(sizes,log=true,bins=bins,alpha=0.2,normed=true,label="dense")
PyPlot.plt.hist(sizes,log=true,bins=bins,alpha=0.2,normed=true,label="sparse")

#PyPlot.plt.hist(sizes_full,log=true,bins=bins,alpha=0.5,normed=true,label="full")
PyPlot.plt.gca()[:set_xscale]("log")

w_range = bins
P_w_th_range = normed_distribution(w_range,P_w_th(w_range,s(sqrt(w_range)./N,alpha,beta)))
P_w_th_range_eff = normed_distribution(w_range,P_w_th(w_range,get_s_eff(sqrt(w_range)./N,alpha,beta,N-1)))

plot(w_range,P_w_th_range,"-r",label="theory")#$P(w) \sim e^{- s(\sqrt{w})^2 w/4} w^{-3/2}/(1 + s(\sqrt{w}))$ (theory)')
plot(w_range,P_w_th_range_eff,"-g",label="effective theory")#$P(w) \sim e^{- s(\sqrt{w})^2 w/4} w^{-3/2}/(1 + s(\sqrt{w}))$ (theory)')


PyPlot.plt.legend()
```

```julia
k = 4

n_n = 5
c_r = 0.5
N = 20
beta = 4.0/(c_r*n_n)
alpha = (N*beta)/n_n
println(N, ' ' ,alpha, ' ',beta)
```

```julia
im = InfectionModel(x -> 1 + alpha*x , x -> 1 + beta)
```

```julia
function normed_distribution(x,px)
    return px./sum(diff(x).*px[1:end-1])
end

P_w_th(w,s) = exp(-s.^2.*w./4).*w.^(-1.5)./(2*sqrt(pi).*(1 .+ s))

s(y,alpha,beta) = alpha*y - beta

get_y_eff(y,k) = y.*(1 + (1-y)./(y*k))

get_s_eff(y,alpha,beta,k) = alpha*get_y_eff(y,k) - beta
```

```julia
sizes,num_fixed = run_epidemics(N,k,10,im)
#sizes_full,num_fixed_full = run_epidemics(N,N-1,20000,im)
```
